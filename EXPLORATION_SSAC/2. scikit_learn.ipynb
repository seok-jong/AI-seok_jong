왜냐하면 예측이 틀린 경우는 해당 클래스를 제외한 나머지 경우(9가지)에 해당하기 때문이다.   
따라서 한 손글씨에 대해 정확히 맞다고 판단하는 것이 중요하다는 결론이다.   
오차 행렬에서 **FN (거짓 부정)**이 중요하다.  
즉, 평가지표상 **Recall**이 중요한 평가지표가 된다고 생각한다.

## 평가지표
유방암 발생의 경우 양성일 때, 음성이라고 판단하면 그 환자는 심각한 위험에 노출된다.  
하지만 음성인 환자가 양성이라 판단을 받는 경우는 위험을 초래하지 않는다.  
따라서 오차행렬에서 **FN (거짓 부정)**이 중요하다고 판단이 되고, **Recall이 중요한 평가지표가 된다고 생각한다.**
